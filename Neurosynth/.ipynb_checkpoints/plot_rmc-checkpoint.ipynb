```python
import numpy as np
import seaborn as sns
import scipy.io as sio
```

```python
%matplotlib inline
sns.set_style("white",{'axes.linewidth': 0.75,})
sns.set_context("poster")
```

```python
probiter=np.array([0.2,0.4,0.6,0.8])
ms_rmc=sio.loadmat('../results/ms_rmc.mat')
msSMC=ms_rmc['msSMC']
msRMC=ms_rmc['msRMC']

```

```python
pl=['NDCG@10',r'Spearman $\rho$',r'Kendall $\tau$']
f,ax=sns.plt.subplots(1,3,figsize=(18,6))
for i in range(len(pl)):
    ax[i].errorbar(probiter,msSMC[:,i,0],yerr=msSMC[:,i,1],fmt='-bo',capthick=3);
    ax[i].errorbar(probiter,msRMC[:,i,0],yerr=msRMC[:,i,1],fmt='-ro',capthick=3);
    ax[i].set_xlim([0,1])
    ax[i].set_ylabel(pl[i]);
    ax[i].set_xlabel(r'$\frac{|\Omega|}{d_1d_2}$',fontsize='24');
#xlabel('Fraction of matrix used in training')
sns.plt.tight_layout()
```
